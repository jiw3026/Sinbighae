```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc("font", family = "NanumBarunGothic")
sns.set(font="NanumBarunGothic", 
rc={"axes.unicode_minus":False}, style='white')
```

```python
data1_gender = data1.sort_values(by='P1',ascending = False)
plt.figure(figsize = (5,5))
plt.title('성별 소비금액',fontsize=15)
ax = sns.countplot(data=data1_gender,x='P1')
for p in ax.patches:
  height = p.get_height()
  ax.text(p.get_x()+p.get_width()/2, height-20,height, ha='center',size=12)
```
